```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
# https://data.world/makeovermonday/2021w30a
filename = '/content/drive/MyDrive/04 Week 4: Explanatory Viz/US Population by Race.xlsx'
df = pd.read_excel(filename)
df
```

```python
df.isnull().sum()
df.fillna({'Native Hawaiian / Other Pacific Islander': 0, 'Asian': 0, 'American Indian / Alaska Native': 0}, inplace=True)
df
```

```python
seriesW = df.groupby('State')['White'].mean()
seriesB = df.groupby('State')['Black'].mean()
seriesH = df.groupby('State')['Hispanic'].mean()
seriesA = df.groupby('State')['Asian'].mean()
```

```python
plot_bar = pd.melt(df, id_vars=['State'], value_vars=['White', 'Black', 'Hispanic', 'Asian'], var_name='Race', value_name='Population')
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.melt.html
plt.figure(figsize=(25,10))
ax = sns.barplot(data=plot_bar, x='State', y='Population',hue='Race')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45,ha='right');
ax.set_title('Race/Nationality Percentage per State', fontsize=18);
```
